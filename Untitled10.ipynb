```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

```

```
path = '/content/activities.csv'
data_raw = pd.read_csv(path)

columns = ['length', 'time']

df = pd.DataFrame(columns=columns)

df['time'] = (data_raw['Beweegtijd']/60).astype('float')
df['length'] = ((data_raw['Afstand']).apply(lambda x: str(x.replace(',','.')))).astype('float')
df['speed'] = (df['time'])/df['length']
```

```
df = df.drop(df[df['length'] > 20].index)
```

```
plt.scatter(df['length'],df['time'], color='#47DBCD', label='Data Point')
plt.legend()
plt.show()
```

```
# initializing our inputs and outputs
X = df['length'].values
Y = df['time'].values

# mean of our inputs and outputs
x_mean = np.mean(X)
y_mean = np.mean(Y)

#number of samples
n = len(X)

# linear regression formula to calculate intercept and coefficient
numerator = 0
denominator = 0
for i in range(n):
    numerator += (X[i] - x_mean) * (Y[i] - y_mean)
    denominator += (X[i] - x_mean) ** 2
b1 = numerator / denominator
b0 = y_mean - (b1 * x_mean)
#printing the coefficient
print(b1, b0)
```

```
x_max = 25 + 1
x_min = np.min(X) - 1

#prediction values
x = np.linspace(x_min, x_max, 1000)
y = b0 + b1 * x

plt.figure(figsize=(12,8))
plt.plot(x, y, color='#F5B14C', label='Linear Regression')
#plot the data point
plt.scatter(X, Y, color='#47DBCD', label='Data Point')
#Half marathon line
plt.axvline(x=21.0975,linestyle=':', label='Distance half marathon' )
# x-axis label
plt.xlabel('Length (km)')
#y-axis label
plt.ylabel('Time (minutes)')
plt.legend()
plt.show()
```

```
pred_half = b0 + b1*21.0975
hours, minutes = divmod(pred_half, 60)

print('Model prediction on time to run half marathon,',"%02d:%02d"%(hours,minutes))
```

```
#Measure of model performance 
rmse = 0
for i in range(n):
    y_pred=  b0 + b1* X[i]
    rmse += (Y[i] - y_pred) ** 2
    
rmse = np.sqrt(rmse/n)
print(rmse)
```

```
sumofsquares = 0
sumofresiduals = 0
for i in range(n) :
    y_pred = b0 + b1 * X[i]
    sumofsquares += (Y[i] - y_mean) ** 2
    sumofresiduals += (Y[i] - y_pred) **2
    
score  = 1 - (sumofresiduals/sumofsquares)
print(score)
```
